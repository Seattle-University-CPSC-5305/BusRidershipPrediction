```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as py


# Read in weather data
w2012 = pd.read_csv('/content/Chicago2012.csv')
w2013 = pd.read_csv('/content/Chicago2013.csv')
w2014 = pd.read_csv('/content/Chicago2014.csv')
w2015 = pd.read_csv('/content/Chicago2015.csv')
w2016 = pd.read_csv('/content/Chicago2016.csv')
w2017 = pd.read_csv('/content/Chicago2017.csv')
w2018 = pd.read_csv('/content/Chicago2018.csv')
w2019 = pd.read_csv('/content/Chicago2019.csv')
```

```python
# Combine all seperate weather datasets with concat
weather = pd.concat([w2012, w2013, w2014, w2015, w2016, w2017, w2018, w2019], ignore_index=True)
```

```python
print(weather.head())
print(weather.info())
```

```python
# Plot attributes to get idea of shape
weather.plot(x = 'DATE', y = 'DAPR')
weather.plot(x = 'DATE', y = 'MDPR')
weather.plot(x = 'DATE', y = 'PRCP')
weather.plot(x = 'DATE', y = 'SNOW')
weather.plot(x = 'DATE', y = 'SNWD')
weather.plot(x = 'DATE', y = 'TAVG')
weather.plot(x = 'DATE', y = 'TMAX')
weather.plot(x = 'DATE', y = 'TMIN')
weather.plot(x = 'DATE', y = 'TOBS')
```

```python
# Use mean() to combine data from all weather stations into average for whole of Chicago
weather_avg = weather.groupby(['DATE'])[['DAPR', 'MDPR', 'PRCP', 'SNOW', 'SNWD', 'TAVG',
                                         'TMAX', 'TMIN', 'TOBS', 'WT01','WT02', 'WT03', 'WT04', 'WT05',
                                         'WT06', 'WT08', 'WT09', 'WT10', 'WT11', 'WT13', 'WT14', 'WT16', 'WT18', 'WT19']].mean()
weather_avg = weather_avg.reset_index()
print(weather_avg.head())
```

```python
weather_avg.rename(columns={'WT01':'fog', 'WT02':'heavy_fog', 'WT03':'thunder',
                            'WT04':'sleet', 'WT05':'hail', 'WT06':'glaze', 'WT08':'smoke',
                            'WT09':'blowing_snow', 'WT10':'tornado', 'WT11':'high_wind',
                            'WT13':'mist', 'WT14':'drizzle', 'WT16':'rain',
                            'WT18':'snow', 'WT19':'PRCP_unknown_source'}, inplace = True)
```

```python
weather_avg.info()
```

```python
# Calculate TAVG for tuples missing this data point
weather_avg.loc[(pd.isnull(weather_avg.TAVG)), 'TAVG'] = weather_avg['TMAX']/2 + weather_avg['TMIN']/2
```

```python
weather_avg.info()
```

```python
# Plot TAVG to check that replacaing missing values did not affect shape of data
plt.hist(weather_avg['TAVG'])
```

```python
# Save new dataset to csv file
weather_avg.to_csv('Chi_Weather_2012_2019.csv', header=True, index=False)
```
